```python
import pre_ms_models_database as md
import pandas as pd
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
# load all data of the grid
grid_all_cols = md.load_data(['6_dim_grid', '10_dim_grid', '10_dim_grid_extension'])
```

```python
# load the individual grids for a quick comparison (will only be used in the cell below)
a = md.load_data(['6_dim_grid'])
b = md.load_data(['10_dim_grid'])
c = md.load_data(['10_dim_grid_extension'])
d = md.load_data(['10_dim_grid', '10_dim_grid_extension'])
```

```python
# print the number of models in the grid (/3 because there are 3 columns per model (for each l value))
# print the number of models for each parent model (number of MESA snapshots for a given star along its evolution)
print("Number of models")
print(f"6_dim_grid: {len(a)/3}")
print(f"10_dim_grid: {len(b)/3}")
print(f"10_dim_grid_extension: {len(c)/3}")
print(f"10_dim_grid_all: {len(d)/3}")
print(f"all: {len(grid_all_cols)/3}")

a_counts = a.groupby("parent_model_numbers").size()/3
b_counts = b.groupby("parent_model_numbers").size()/3
c_counts = c.groupby("parent_model_numbers").size()/3
d_counts = d.groupby("parent_model_numbers").size()/3

print("Mean, median, std")
print(f"6_dim_grid: {a_counts.mean()}, {a_counts.median()}, {a_counts.std()}")
print(f"10_dim_grid: {b_counts.mean()}, {b_counts.median()}, {b_counts.std()}")
print(f"10_dim_grid_extension: {c_counts.mean()}, {c_counts.median()}, {c_counts.std()}")
print(f"10_dim_grid_all: {d_counts.mean()}, {d_counts.median()}, {d_counts.std()}")
```

```python
# these columns are not needed for the analysis
drop_cols = [
        "F_drot",
        "O1_drot",
        "O2_drot",
        "O3_drot",
        "O4_drot",
        "O5_drot",
        "O6_drot",
        "O7_drot",
        "O8_drot",
        "O9_drot",
        "parent_model_numbers",
        "index"
    ]
```

```python
grid = grid_all_cols.drop(columns=drop_cols)
```

```python
grid
```

```python
# in the following we flatten the l values to have a single row per model
pivot_index = [
    "model_id",
    "mass",
    "Teff",
    "log_g",
    "radius",
    "L",
    "age",
    "alphas",
    "zs",
    "e_fold",
    "betas",
    "mlins",
    "mdots",
    "Dmixs",
    "f1s",
    "hes",
]
pivot_values = ["F", "O1", "O2", "O3", "O4", "O5", "O6", "O7", "O8", "O9"]
```

```python
# Pivot the DataFrame
grid_pivot = grid.pivot(
    index=pivot_index,
    columns="l_val",
    values=pivot_values,
)
```

```python
grid_pivot
```

```python
# Flatten the multi-level columns
grid_pivot.columns = [f"{col[0]}_l_{int(col[1])}" for col in grid_pivot.columns]
```

```python
grid_pivot
```

```python
grid_concatted = grid_pivot.reset_index().set_index('model_id')
```

```python
# remove rows that contain zero values (only affects the frequencies; these are not physical models and are caused by issues in the GYRE calculations)
grid_concatted = grid_concatted[(grid_concatted != 0).all(axis=1)]
```

```python
# split the data into input and output columns (fundamental and observable parameters)
output_cols = [
    "Teff",
    "log_g",
    "radius",
    "L",
    "F_l_0",
    "F_l_1",
    "F_l_2",
    "O1_l_0",
    "O1_l_1",
    "O1_l_2",
    "O2_l_0",
    "O2_l_1",
    "O2_l_2",
    "O3_l_0",
    "O3_l_1",
    "O3_l_2",
    "O4_l_0",
    "O4_l_1",
    "O4_l_2",
    "O5_l_0",
    "O5_l_1",
    "O5_l_2",
    "O6_l_0",
    "O6_l_1",
    "O6_l_2",
    "O7_l_0",
    "O7_l_1",
    "O7_l_2",
    "O8_l_0",
    "O8_l_1",
    "O8_l_2",
    "O9_l_0",
    "O9_l_1",
    "O9_l_2",
]
```

```python
inputs = grid_concatted[
    [
        "mass",
        "age",
        "alphas",
        "zs",
        "e_fold",
        "betas",
        "mlins",
        "mdots",
        "Dmixs",
        "f1s",
        "hes",
    ]
]
```

```python
outputs = grid_concatted[output_cols]
```
