```python
# Define the modelling_ridge class with delta_nu calculation
class modelling_ridge():
    def __init__(self, f_dict, l): 
        self.l = l  # store the spherical mode degree l = 0, 1, 2
        
        # Extract and store frequency values and we ignore NaN values!!
        self.fvals = np.array([f_dict[key] for key in f_dict if pd.notna(f_dict[key])])
        
        # Compute delta_nu based on frequency separations, median distance of radial modes
        if len(self.fvals) > 1: # only do this if there is more than one frequency!!
            self.delta_nu_vals = self.fvals[1:] - self.fvals[:-1]
            self.delta_nu = np.median(self.delta_nu_vals)  
        else:
            self.delta_nu_vals = []
            self.delta_nu = 0  # Prevents errors if there's only one frequency
        
        self.mod_fvals = self.fvals % self.delta_nu 

def calculate_delta_nu_for_model(row):
    # Extract only l=0 frequencies
    l0_columns = [col for col in row.index if col.endswith("_l_0")]
    
    # Extract corresponding frequency values
    fvals = {col: row[col] for col in l0_columns if pd.notna(row[col])}
    all_fvals = sorted([value for value in fvals.values() if pd.notna(value)])  # Sort frequencies

    # Only calculate delta_nu if at least two frequencies exist, otherwise it doesn't make sense at all
    if len(all_fvals) > 1:
        delta_nu_vals = np.diff(all_fvals)  # Compute frequency separations
        return np.median(delta_nu_vals)  # Return median frequency separation
    
    return 0 # if insufficient data


# this is where we add the delta nu column to our previous dataset grid_pivotnozeroes not containing the delta nu values YET 
# and although we recalculate the delta nu values in the cell below, when we plot the ridges, THIS is the method to calculate the delta_nus
grid_concatted.loc[:, 'delta_nu'] = grid_concatted.apply(calculate_delta_nu_for_model, axis=1)

print(grid_concatted.head())
```

```python
random_model = grid_concatted.sample(n=1).iloc[0]
delta_nu = random_model["delta_nu"]
ridges = []
print(f"Global delta_nu: {delta_nu}")  # Print global delta_nu from the model

# Function to plot ONE echelle diagram 
def plot_echelle(ridges, dnu):
    fig, ax = plt.subplots(figsize=(8, 6), dpi = 130)
    colors = {0: "darkorange", 1: "mediumblue", 2: "deeppink"}
    markers = {0: "o", 1: "s", 2: "D"}

    # Here we just loop through the ridges and plot them
    for r in ridges:
        ax.scatter(r.mod_fvals, r.fvals, c=colors[r.l], marker=markers[r.l], 
                   edgecolors='black', alpha=0.6, s=70, label=f'l={r.l}')
        ax.scatter(r.mod_fvals+dnu, r.fvals, c=colors[r.l], marker=markers[r.l], 
                   edgecolors='black', alpha=0.6, s=70)
    ax.axvline(dnu, color='black', ls='--', lw=2)
    ax.set_ylabel('Frequency $(d^{-1})$')
    ax.set_xlabel(f'Frequency modulo {dnu:.3f} $(d^{-1})$')
    ax.set_xlim(0, dnu*1.2)
    ax.legend()
    plt.show()

# Loop over each l value (0, 1, 2) and calculate the delta_nu for the selected model
for l in [0, 1, 2]:
    # Get columns corresponding to current l value
    l_columns = [col for col in random_model.index if col.endswith(f"_l_{l}")]
    
    # Extract frequency values for the current l value
    fvals = {col: random_model[col] for col in l_columns}
    all_fvals = [value for value in fvals.values() if pd.notna(value)]  # Filter NaN values
    
    # Only consider ridges if there are at least two valid frequencies
    if len(all_fvals) > 1:  # Now using the actual values (not modulo yet)
        ridge = modelling_ridge(fvals, l)  # Create the ridge object
        ridges.append(ridge)  # Add the ridge object to the list

# Now, plot the echelle diagram for the random model!

if ridges:
    print(f"Delta nu calculated with modelling_ridges: {ridges[0].delta_nu}")  # Check delta_nu calculated with modelling_ridges
    plot_echelle(ridges, delta_nu)  # Plot the echelle diagram, here we could also replace ridges[0].delta_nu with delta_nu
    # but I think we need the above for loop in order to create the ridges for this particular case
else:
    print("Not enough valid frequencies to create an echelle diagram.")
```
